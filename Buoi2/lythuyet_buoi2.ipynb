## II. Kiểu dữ liệu của biến

### 1. Các kiểu dữ liệu
Có nhiều kiểu dữ liệu khác nhau trong Python. Một số loại quan trọng được liệt kê dưới đây:
- ***Kiểu văn bản:***
    - **str:** chuỗi ký tự (e.g. "hello")
- ***Kiểu số:***
    - **int:** số nguyên (e.g 1, -5)
    - **float:** số thực (e.g. 3.14, -0.5)
    - **complex:** số phức (e.g. 3+5j, -5j)

- ***Kiểu có cấu trúc:***
    - **list, tuple, set, dict:** cấu trúc dữ liệu phức tạp (sẽ được học ở bài sau)

### 6. Biểu thức

#### Toán tử

##### 1. Toán tử số học

| Toán tử | Ý nghĩa             | Ví dụ       |
|---------|---------------------|-------------|
| `+`     | Cộng                | `x + y`     |
| `-`     | Trừ                 | `x - y`     |
| `*`     | Nhân                | `x * y`     |
| `/`     | Chia                | `x / y`     |
| `//`    | Chia lấy phần nguyên| `x // y`    |
| `%`     | Chia lấy dư         | `x % y`     |
| `**`    | Lũy thừa            | `x ** y`    |

##### 2. Toán tử gán (Assignment Operators)

| Toán tử | Ý nghĩa                    | Ví dụ        |
|---------|----------------------------|--------------|
| `=`     | Gán giá trị                | `x = 5`      |
| `+=`    | Cộng rồi gán               | `x += 3`     |
| `-=`    | Trừ rồi gán                | `x -= 2`     |
| `*=`    | Nhân rồi gán               | `x *= 4`     |
| `/=`    | Chia rồi gán               | `x /= 2`     |
| `//=`   | Chia nguyên rồi gán        | `x //= 2`    |
| `%=`    | Chia dư rồi gán            | `x %= 3`     |
| `**=`   | Lũy thừa rồi gán           | `x **= 2`    |

##### 3. Toán tử so sánh (Comparison Operators)

| Toán tử | Ý nghĩa                     | Ví dụ        |
|---------|-----------------------------|--------------|
| `==`    | So sánh bằng                | `x == y`     |
| `!=`    | Khác nhau                   | `x != y`     |
| `>`     | Lớn hơn                     | `x > y`      |
| `<`     | Nhỏ hơn                     | `x < y`      |
| `>=`    | Lớn hơn hoặc bằng           | `x >= y`     |
| `<=`    | Nhỏ hơn hoặc bằng           | `x <= y`     |

##### 4. Toán tử logic (Logical Operators)

| Toán tử | Ý nghĩa                     | Ví dụ             |
|---------|-----------------------------|--------------------|
| `and`   | Và – đúng khi cả hai đúng   | `x > 5 and x < 10` |
| `or`    | Hoặc – đúng nếu một trong hai đúng | `x < 5 or x > 10` |
| `not`   | Phủ định – đảo ngược kết quả | `not(x > 5)`      |

##### 5. Toán tử thành viên (Membership Operators)

| Toán tử | Ý nghĩa                         | Ví dụ             |
|---------|----------------------------------|--------------------|
| `in`    | Kiểm tra phần tử có trong chuỗi, list... | `'a' in 'abc'`    |
| `not in`| Kiểm tra phần tử không có trong  | `'z' not in 'abc'`|

##### 6. Toán tử nhận dạng (Identity Operators)

| Toán tử | Ý nghĩa                                | Ví dụ         |
|---------|-----------------------------------------|---------------|
| `is`    | Kiểm tra 2 biến có cùng đối tượng không | `x is y`      |
| `is not`| Kiểm tra 2 biến không cùng đối tượng    | `x is not y`  |

---

##### 7. Toán tử bit (Bitwise Operators)

| Toán tử | Ý nghĩa               | Ví dụ    |
|---------|------------------------|----------|
| `&`     | AND                    | `x & y`  |
| `\|`    | OR                     | `x \| y` |
| `^`     | XOR                    | `x ^ y`  |
| `~`     | NOT                    | `~x`     |
| `<<`    | Dịch trái              | `x << 2` |
| `>>`    | Dịch phải              | `x >> 2` |

## III. Cấu trúc điều khiển
### 1. Cấu trúc rẽ nhánh: *if*
#### Cấu trúc *if:*
```
if<condition>:
    command
```
#### Cấu trúc *if...else:*
```
if<condition>:
    command1
else:
    command2
```
#### Cấu trúc *if...elif...else:*
```
if<condition_1>:
    command1
elif<condition_2>:
    command2
else:
    command3
```
Trong đó:
- `<condition>`, `<condition_1>`, `<condition_2>`: Là một biểu thức logic
- `command1, command2, command3`: có thể là 1 lệnh, một cấu trúc điều khiển, một khối lệnh, hoặc khối cấu trúc điều khiển

````python
x = 5
if x == 0:
    print("Số không")
elif x > 0:
    print("Số dương")
elif x < 0:
    print("Số âm")
else:
    print("Không phải số")
````

### 2. Vòng lặp for
Vòng lặp for trong Python được sử dụng để lặp qua một chuỗi (ví dụ như danh sách, tuple, chuỗi ký tự) hoặc các đối tượng có thể lặp khác. Lặp theo một trình tự được gọi là duyệt.

**Cú pháp:**
```
for <variable> in <array>:
    command
```
**Trong đó:**
 - **<Biến>** là một tên do ta tùy ý đặt. Biến sẽ lần lượt nhận từng giá trị trong
<Dãy> và thực hiện Lệnh_lặp.
 - **Lệnh_lặp:** có thể là 1 lệnh, một khối lệnh, một cấu trúc điều khiển hoặc một
khối cấu trúc điều khiển

````python
for x in (1, 2, 5):
    print(x)
````

### 3. Hàng range()

Hàm range() thường được dùng kèm với lệnh for khi nó cho phép ta khởi tạo một dãy với nhiều dạng khác nhau:


````python
 # `range(n)`: Khởi tạo các giá trị nguyên liên tiếp, từ 0 tới n - 1.
for x in range(5):
    print(x)
````

````python
# range(n, m): Khởi tạo các giá trị nguyên liên tiếp, từ n tới m - 1.
for x in range(2, 5):
    print(x)
````

````python
# range(start, stop, step) Khởi tạo một dãy các số nguyên từ start cho tới stop-1, nhưng với bước nhảy step.
for x in range(2, 10, 3):
    print(x)
````

### 4. Vòng lặp while
Vòng lặp while trong Python được sử dụng để lặp qua một lệnh/ một khối lệnh miễn là biểu thức điều kiện là đúng. Chúng ta thường sử dụng vòng lặp này khi chúng ta không biết trước số lần lặp lại.

**Cú pháp:**
```
while <điều_kiện_lặp>:
    <Lệnh_lặp>
```

````python
x = 0
while x < 5:
    print(x)
    x += 1
````
